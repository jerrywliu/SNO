# 2D Kuznetsov Ma

For Chebyshev coefficients.

```python
N_points = 100
periodic = False
data = Breather.breather(N_points, N_points, 2, key=random.PRNGKey(12), periodic=periodic, return_coefficients=True)
if periodic:
  x = jnp.hstack([utils.grid(N_points, periodic=True), 1])
else:
  x = utils.grid(N_points, periodic=False)
```

```python
plt.yscale("log")
plt.plot(jnp.sum(abs(data[0][1][:, :]), axis=1), "o", color="red")
plt.plot(jnp.sum(abs(data[0][1][:, :]), axis=0), "o", color="black");
```

```python
norm = jnp.linalg.norm(data[0][1][:, :])
errors = [jnp.linalg.norm(data[0][1][:i, :i]) / norm for i in range(1, 30)]
plt.yscale("log")
plt.plot(range(1, 30), errors, "o", color="black");
```

```python
plt.yscale("log")
plt.plot(abs(data[1][1][:, 0]), "o", color="red")
plt.plot(abs(data[1][1][0, :]), "o", color="black");
```

```python
norm = jnp.linalg.norm(data[1][1][:, :])
errors = [jnp.linalg.norm(data[1][1][:i, :i]) / norm for i in range(1, 30)]
plt.yscale("log")
plt.plot(range(1, 30), errors, "o", color="black");
```
