```python
from datasets import Square_shift
from functions import utils
```

Square of trigonometric series $f(x) \longrightarrow f(x)f(x+1)$

```python
N_points = 100

data = Square_shift.Fourier_series(10, N_points, 5, key=random.PRNGKey(13), periodic=True, return_coefficients=False)
x = utils.grid(N_points, periodic=True)
```

```python
plt.plot(x, data[0][10], color="black")
plt.plot(x, data[1][10], color="orange");
```

$L_2$ norm $=1$

```python
jnp.linalg.norm(data[0][0])*(x[1] - x[0])**(1/2)
```

Can also evaluate the same function on the Chebyshev grid

```python
N_points = 100

data = Square_shift.Fourier_series(10, N_points, 5, key=random.PRNGKey(13), periodic=False, return_coefficients=False)
x = utils.grid(N_points, periodic=False)
```

```python
plt.plot(x, data[0][0], color="black")
plt.plot(x, data[1][0], color="orange");
```

In the space of Chebyshev coefficients

```python
N_points = 100

data = Square_shift.Fourier_series(15, N_points, 5, key=random.PRNGKey(13), periodic=False, return_coefficients=True)
```

```python
plt.yscale("log")
plt.plot(abs(data[0][0]), "o", color="black")
plt.plot(abs(data[1][0]), "o", color="orange");
```

In the space of Fourier coefficients.

```python
N_points = 100

data = Square_shift.Fourier_series(15, N_points, 5, key=random.PRNGKey(13), periodic=True, return_coefficients=True)
```

```python
plt.yscale("log")
plt.plot(abs(data[0][0]), "o", color="black")
plt.plot(abs(data[1][0]), "o", color="orange");
```
