Solution to the ODE

\begin{equation}
  \frac{d}{dt}y(t) = y(t)\sum_{k}\exp(\pi k i t)c_{k}
\end{equation}

```python
N_points = 100

data = Parametric_ODE.Fourier_series(30, N_points, 5, key=random.PRNGKey(13), shift=1, periodic=False, return_coefficients=False)
x = utils.grid(N_points, periodic=False)
```

```python
N_sample = 0
plt.plot(x, data[0][N_sample], color="black")
plt.plot(x, data[1][N_sample], color="orange");
```
